## Area codes with å, ä, ö to population data

```python
population_df = pd.read_csv('../data/population_df.csv')

# Join municipality, code and name to form area code, add prefix 0
population_df['Area no-dots'] = '0' + population_df['municipality'].astype(str) + " " + population_df['code'].astype(str) + " " + population_df['name']

# Get original area names with letters å, ä, ö:
# Fetch district data
districts = pd.read_csv('../data/Helsinki_districts.csv')
districts = districts[districts['BASIC'] != 0]
districts = districts[['Helsinki statistical area code', 'Helsinki statistical area code 2']]
# Change 091 801 Ostersundomin peruspiiri to 091 801 ostersundomin peruspiiri
districts['Helsinki statistical area code 2'] = districts['Helsinki statistical area code 2'].str.replace('091 801 Ostersundomin peruspiiri', '091 801 ostersundomin peruspiiri') 

# Merge district names with letters å, ä, ö to population_df
population_df = pd.merge(left=population_df, right=districts, how='left', left_on='Area no-dots', right_on='Helsinki statistical area code 2')
population_df = population_df.drop('Helsinki statistical area code 2', axis=1)
population_df['Helsinki statistical area code'] = population_df['Helsinki statistical area code'].fillna(population_df['Area no-dots'])
population_df = population_df.rename(columns={'Helsinki statistical area code': 'Area'})

population_df.tail(20)
```

```python
# Save to file population_df2.csv
population_df.to_csv('../data/population_df2.csv', index=False)
```

## Data for one-person households

```python
# Fetch data from file in folder 'data'
ones = pd.read_csv('../data/Helsinki_oneperson_households.csv')
ones
```

```python
# Get data for total population per gender per district

# Fetch data of all population per district
allpeople = pd.read_csv('../data/Helsingin_vaesto_sukupuolen_ian_ja_aidinkielen_ruotsinkieliset_mukaan.csv',
                       encoding='ISO-8859-1')

# Remove rows that include total data for both genders ('Molemmat sukupuolet')
allpeople = allpeople[allpeople['Sukupuoli'] != 'Molemmat sukupuolet']
# Keep only needed columns
allpeople = allpeople[['Alue', 'Sukupuoli', 'Väestö yhteensä 2022']]
# Replace missing population information '..' with NaN
allpeople['Väestö yhteensä 2022'] = allpeople['Väestö yhteensä 2022'].replace('..', np.nan)


# Fetch district data and convert to same format as in one-person household data
distr = pd.read_csv('../data/Helsinki_districts.csv')
distr = distr[distr['BASIC'] != 0]

distr[['City code', 'District code', 'District level name']] = distr['Helsinki statistical area code'].str.split(' ', n=2, expand=True)
distr['District'] = distr['District code'] + " " + distr['Name']
distr = distr[['Helsinki statistical area code', 'District']]


# Join district code & name with allpeople
people = pd.merge(left=allpeople, right=distr, how='inner', left_on='Alue', right_on='Helsinki statistical area code')
people = people[['Sukupuoli', 'Väestö yhteensä 2022', 'District', 'Alue']]
people.columns = ['Gender', 'Population', 'District', 'District long code']
people['Gender'] = people['Gender'].replace({'Miehet': 'Men', 'Naiset': 'Women'})

people.head(50)
```

```python
# Join total population data with one-person household data
ones2 = pd.merge(left=ones, right=people, how='inner', left_on=['One-person hh, District', 'One-person hh, Gender'], right_on=['District', 'Gender'])
ones2 = ones2.drop(['One-person hh, District', 'One-person hh, Gender', 'One-person hh, Major district'], axis=1)
ones2['Population'] = ones2['Population'].apply(pd.to_numeric)
ones2
```

```python
# Change the age specific columns from amounts to portions of total population

ones_portions = ones2.copy()

# Divide amounts of people in each one-person hh column by total population in district
# and add suffix '%'' to column names to indicate portions
cols_include = ones_portions.columns[~ones_portions.columns.isin(['Gender', 'Population', 'District', 'District long code'])]
for col in cols_include:
    ones_portions[col] = ones2[col].div(ones2['Population'], axis=0)
    ones_portions.rename(columns={col: col+', %'}, inplace=True)

# Merge with sub-district-level population data
ones_portions = pd.merge(left=ones_portions, right=people, how='right', 
                         left_on=['District', 'Gender'], right_on=['District', 'Gender'])
ones_portions = ones_portions.drop(['Population_x', 'District'], axis=1)
ones_portions = ones_portions.rename(columns={'District long code_x': 'District',
                                              'District long code_y': 'Area',
                                              'Population_y' : 'Population'})

# Fill down district-level name
ones_portions['District'] = ones_portions['District'].ffill()


# Copy rows where population is missing
no_pop = ones_portions[ones_portions['Population'].isna()]

# Remove the rows from ones_portions 
ones_portions = ones_portions[~ones_portions['Population'].isna()]

# Sort rows based on District, Gender and All ages
ones_portions = ones_portions.sort_values(by = ['District', 'Gender', 'One-person hh, All ages, %'], ascending = [True, True, False])
 
# Fill down values in portions columns to sub-district level
ones_portions = ones_portions.ffill()

# Append no-population rows back to population data
ones_portions =  pd.concat([ones_portions, no_pop])

# remove columns District and Population
ones_portions = ones_portions.drop(['District', 'Population'], axis=1)


# Change gender to categorical type
ones_portions['Gender'] = ones_portions['Gender'].map({'Men':0, 'Women':1})
ones_portions['Gender'] = ones_portions['Gender'].astype('category') 


# Sort along index
ones_portions.sort_index(inplace=True)


ones_portions
```

### Merge one-person housefolds with population dataframe

```python
# Read population data (including district names with and without å, ä, ö)
population_df2 = pd.read_csv('../data/population_df2.csv')

# Change gender to categorical type
population_df2['Gender'] = population_df2['Gender'].astype('category') 

population_df2
```

```python
# Merge one-person households with population data
population_df3 = pd.merge(left=population_df2, right=ones_portions, how='left', on=['Gender', 'Area'])

population_df3
```

```python
# Save to file population_df3.csv
population_df3.to_csv('../data/population_df3.csv', index=False)
```
